```python
import numpy as np
import pandas as pd
```

```python
def sales_amount(prod_desc,unit,sales_qty):
  if(unit == 'pcs'):
    arr = prod_desc.split("-")
    return float(arr[1])*sales_qty
  else:
    return sales_qty
```

```python
df = pd.read_csv("/content/main.csv")
```

```python
df
```

```python
df['sales_amount'] = df.apply(lambda row : sales_amount(row['product_description'],row['unit'],row['sales_quantity']) , axis = 1)
```
